```python
# Install (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
import numpy as np
from matplotlib import pyplot as plt
import urllib.request

# -------------------------------
# 1. Use a stable image URL
# -------------------------------
image_url = "https://upload.wikimedia.org/wikipedia/commons/3/3a/Cat03.jpg"

# Add User-Agent header to avoid 403 Forbidden error
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
urllib.request.install_opener(opener)

urllib.request.urlretrieve(image_url, "sample.jpg")

# -------------------------------
# 2. Read image
# -------------------------------
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
    exit()

# -------------------------------
# 3. Apply Dilation
# -------------------------------
kernel = np.ones((5,5), np.uint8)
dilated = cv2.dilate(img, kernel, iterations=1)

# -------------------------------
# 4. Convert for display
# -------------------------------
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dilated_rgb = cv2.cvtColor(dilated, cv2.COLOR_BGR2RGB)

# -------------------------------
# 5. Show images
# -------------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Dilated Image")
plt.imshow(dilated_rgb)
plt.axis('off')

plt.show()
```
