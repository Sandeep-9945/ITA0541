```python
# Install (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
from matplotlib import pyplot as plt
import urllib.request
from urllib.request import Request, urlopen

# -------------------------------
# 1. Download sample image
# -------------------------------
image_url = "https://upload.wikimedia.org/wikipedia/commons/3/3a/Cat03.jpg"
# Add a User-Agent header to avoid 403 Forbidden error
req = Request(image_url, headers={'User-Agent': 'Mozilla/5.0'})
with urlopen(req) as response:
    with open("sample.jpg", "wb") as out_file:
        out_file.write(response.read())

# -------------------------------
# 2. Read image
# -------------------------------
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
    exit()

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------------------------------
# 3. Rotate image by 180 degrees
# -------------------------------
rotated = cv2.rotate(img, cv2.ROTATE_180)

rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)

# -------------------------------
# 4. Display images
# -------------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("180 Degree Rotated Image")
plt.imshow(rotated_rgb)
plt.axis('off')

plt.show()
```
