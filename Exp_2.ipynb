```python
# Install libraries (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
from matplotlib import pyplot as plt
import urllib.request

# -------------------------------
# 1. Download sample image
# -------------------------------
image_url = "https://images.unsplash.com/photo-1543466835-00a7907e9de1?auto=format&fit=crop&w=800&q=80"
urllib.request.urlretrieve(image_url, "sample.jpg")

# -------------------------------
# 2. Read image
# -------------------------------
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
    exit()

# Convert original image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------------------------------
# 3. Apply Gaussian Blur
# -------------------------------
blur = cv2.GaussianBlur(img, (15,15), 0)

# Convert blurred image to RGB
blur_rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

# -------------------------------
# 4. Display images
# -------------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Gaussian Blurred Image")
plt.imshow(blur_rgb)
plt.axis('off')

plt.show()
```
