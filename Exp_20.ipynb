```python
# Install libraries (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
import numpy as np
from matplotlib import pyplot as plt
import urllib.request

# -------------------------------
# 1. Download sample image
# -------------------------------
image_url = "https://images.unsplash.com/photo-1564349683136-77e08dba1ef7?auto=format&fit=crop&w=900&q=80"
urllib.request.urlretrieve(image_url, "sample.jpg")

# -------------------------------
# 2. Read image
# -------------------------------
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
    exit()

# Convert image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------------------------------
# 3. Create Kernel
# -------------------------------
kernel = np.ones((5,5), np.uint8)

# -------------------------------
# 4. Apply Dilation
# -------------------------------
dilated = cv2.dilate(img, kernel, iterations=1)

# Convert dilated image to RGB
dilated_rgb = cv2.cvtColor(dilated, cv2.COLOR_BGR2RGB)

# -------------------------------
# 5. Display images
# -------------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Dilated Image")
plt.imshow(dilated_rgb)
plt.axis('off')

plt.show()
```
