```python
# Install (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
from matplotlib import pyplot as plt
import urllib.request

# -------------------------------
# 1. Download sample image
# -------------------------------
image_url = "https://images.unsplash.com/photo-1498050108023-c5249f4df085?auto=format&fit=crop&w=800&q=80"
urllib.request.urlretrieve(image_url, "sample.jpg")

# -------------------------------
# 2. Read image
# -------------------------------
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
    exit()

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------------------------------
# 3. Rotate 270° clockwise
# (same as 90° counterclockwise)
# -------------------------------
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)

# -------------------------------
# 4. Display images
# -------------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("270° Clockwise Rotated")
plt.imshow(rotated_rgb)
plt.axis('off')

plt.show()
```
