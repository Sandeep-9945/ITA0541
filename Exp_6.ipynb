```python
# Install (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load a sample image directly from OpenCV samples
image_url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"

import urllib.request
urllib.request.urlretrieve(image_url, "sample.jpg")

# Read image
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
else:
    # Create kernel
    kernel = np.ones((5,5), np.uint8)

    # Apply erosion
    eroded = cv2.erode(img, kernel, iterations=1)

    # Convert BGR to RGB for matplotlib display
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    eroded_rgb = cv2.cvtColor(eroded, cv2.COLOR_BGR2RGB)

    # Display images
    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.title("Original Image")
    plt.imshow(img_rgb)
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.title("Eroded Image")
    plt.imshow(eroded_rgb)
    plt.axis('off')

    plt.show()
```
