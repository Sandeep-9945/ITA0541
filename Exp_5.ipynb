```python
# Install libraries (only needed in Colab)
!pip install opencv-python matplotlib

import cv2
from matplotlib import pyplot as plt
import urllib.request

# -------------------------------
# 1. Download sample image
# -------------------------------
# Updated the image URL to a direct link from a public GitHub repository for better stability
image_url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
urllib.request.urlretrieve(image_url, "sample.jpg")

# -------------------------------
# 2. Read image
# -------------------------------
img = cv2.imread("sample.jpg")

if img is None:
    print("Error loading image")
    exit()

# Convert image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------------------------------
# 3. Function to analyze histogram
# -------------------------------
def analyze_histogram(image):

    colors = ('b', 'g', 'r')

    plt.figure(figsize=(10,5))

    for i, color in enumerate(colors):

        hist = cv2.calcHist([image], [i], None, [256], [0,256])

        plt.plot(hist, color=color)

    plt.title("Color Histogram")
    plt.xlabel("Pixel Intensity Value")
    plt.ylabel("Number of Pixels")

    plt.show()

# -------------------------------
# 4. Display original image
# -------------------------------
plt.figure(figsize=(5,5))

plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.show()

# -------------------------------
# 5. Analyze histogram
# -------------------------------
analyze_histogram(img)
```
